```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_excel("/content/population_distribution_dataset.xlsx",sheet_name="Population_Data")
```

```
gender_counts = df['Gender'].value_counts()
print(gender_counts)
```

```
plt.figure(figsize=(6,5))


plt.bar(gender_counts.index, gender_counts.values)


plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")

plt.show()
```
